```python
haiti = df_can.loc['Haiti', years] # passing in years 1980 - 2013 to exclude "total" column
haiti.head()
```

```python
haiti.plot()
```

```python
haiti.index = haiti.index.map(int)
haiti.plot(kind='line', c='b')

plt.title('Immigration from Haiti')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')

plt.show()
```

```python
df_can.sort_values(by='Total', ascending=False, axis=0, inplace=True)

df_top5 = df_can.head(5)

df_top5 = df_top5[years].transpose()

df_top5.index = df_top5.index.map(int)

df_top5.plot(kind='line', figsize=(14,8))

haiti.index = haiti.index.map(int)
haiti.plot(kind='line')

plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')

plt.show()
```

# Area Plots

```python
df_can.sort_values(['Total'], ascending=False, axis=0, inplace=True)

df_top5 = df_can.head()

df_top5 = df_top5[years].transpose()

df_top5.head()
```

```python
df_top5.index = df_top5.index.map(int)

df_top5.plot(kind='area', stacked=False, figsize=(20, 10),)

plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

```python
df_top5.plot(kind='area', alpha=0.25, stacked=False, figsize=(20, 10),)

plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

# Two types of plotting


```python
df_top5.plot(kind='area', alpha=0.35, figsize=(20, 10))

plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

```python
df_least5 = df_can.tail(5)
df_least5 = df_least5[years].transpose()

df_least5.plot(kind='area',
              alpha=0.45,
              figsize=(20,10),
              )

plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

```python
ax = df_least5.plot(kind='area', alpha=0.35, figsize=(20, 10))

ax.set_title('Immigration Trend of Top 5 Countries')
ax.set_ylabel('Number of Immigrants')
ax.set_xlabel('Years')
```

# Bar Charts

# Scatter Plots

```python
df_tot = pd.DataFrame(df_can[years].sum(axis=0))

df_tot.index = map(int, df_tot.index)

df_tot.reset_index(inplace = True)

df_tot.columns = ['year', 'total']

df_tot.head
```

```python
df_tot.plot(kind='scatter', x='year', y='total', figsize=(10,6),
            color='darkblue')

plt.title('Total Immigration to Canada from 1980 - 2013')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')

plt.show()
```
